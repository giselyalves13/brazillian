```python
sellers = pd.read_csv("olist_sellers_dataset.csv", ',')
```

```python
itens_sellers = itens.join(sellers, how='inner', lsuffix='_caller', rsuffix='_other')
```

```python
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical
from keras.layers import Dropout
```

```python
Y_itens_sellers = itens_sellers['price']
```

```python
cols = ['shipping_limit_date','freight_value','seller_city','seller_state']
# cols = ['freight_value']

X_itens_sellers = itens_sellers[cols]
```

```python
embed_dim = 128
batch_size= 32

##Buidling the LSTM network

model = Sequential()
model.add(Dense(128,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(128,activation='softmax'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
X_itens_sellers['seller_city'] = X_itens_sellers['seller_city'].astype('category')
X_itens_sellers['seller_state'] = X_itens_sellers['seller_state'].astype('category')
X_itens_sellers['shipping_limit_date'] = X_itens_sellers['shipping_limit_date'].astype('category')

X_itens_sellers['seller_city'] = X_itens_sellers['seller_city'].cat.codes
X_itens_sellers['seller_state'] = X_itens_sellers['seller_state'].cat.codes
X_itens_sellers['shipping_limit_date'] = X_itens_sellers['shipping_limit_date'].cat.codes
```

```python
X_itens_sellers[:10]
```

```python
Y_itens_sellers[:10]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_itens_sellers.values,Y_itens_sellers.values, test_size = 0.20)
```

```python
X_train[:10]
```

```python
Y_train[:10]
```

```python
from keras.callbacks import EarlyStopping
#set early stopping monitor so the model stops training when it won't improve anymore
early_stopping_monitor = EarlyStopping(patience=3)

history = model.fit(X_train, Y_train, validation_split=0.2, epochs=30, callbacks=[early_stopping_monitor])

# Measuring score and accuracy on validation set
# score,acc = model.evaluate(X_test, Y_test, verbose = 1, batch_size = batch_size)
# print("Loss score: %.2f" % (score))
# print("Test Accuracy: %.2f" % (acc*100))
```

```python
model.evaluate(X_test, Y_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
test_y_predictions = model.predict(X_test)
print("X=%s, Predicted=%s, %s" % (X_test[25], Y_test[25], test_y_predictions[25]))
```
